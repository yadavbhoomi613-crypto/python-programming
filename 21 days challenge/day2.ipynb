# FINDING LARGEST NUMBER IN A LIST

```python
n=57
lst=[1,3,5,7,9]
for i in range(n):
    num=10
    lst.append(num)
    largest=lst[0]
    for b in lst:
        if b>largest:
            largest=b
    print("the largest number is:",largest);
```
