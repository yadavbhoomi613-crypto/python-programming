# 1. Take two numbers from user and print after swapping

```python
a=int(input("Enter a number: "))
b=int(input("enter a number:"))
print("before swapping a,b =",a,b)
print("after swapping a,b=",b,a)
```

# 2. Print the sum of first n natural nummbers

```python
n=int(input("enter the first n natural numbers:"))
sum=n*(n+1)//2
print("the sum of first n natural numbers is:",sum)
```

# 3. Find the area of a triangle

```python
s1=float(input("enter the side1 of triangle:"))
s2=float(input("enter the side2 of triangle:"))
s3=float(input("enter the side3 of triangle:")) 
s=(s1+s2+s3)/2
area=(s*(s-s1)*(s-s2)*(s-s3))**0.5
print("the area of triangle is:",area)
```

# 4. Find the SIMPLE INNTEREST

```python
p=float(input("enter the principal amount:"))
r=float(input("enter the rate:"))
t=float(input("enter the time in years:"))
si=(p*r*t)/100 
print("the simple interest is:",si)
```

# 5. Find the area of a circle

```python
r=float(input("enter the radius of circle:"))
area=3.14*r*r
print("the area of circle is:",area)
```
