# 1. Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem

```python
a=float(input("enter base:"))
b=float(input("enter height:"))
c=(a**2+b**2)**0.5
print("the hypotaneuse is:",c)
```

# 2. Write a program to convert given seconds into hours, minutes and remaining seconds.

```python
seconds=float(input("enter the seconds:"))

hours=seconds//3600
minutes=seconds%3600//60
remsecs=seconds%60

print("hours:",hours)
print("minutes:",minutes)
print("seconds:",remsecs)
```

# 3. Using membership operator find whether a given number is in sequence (10,20,56,78,89)

```python
seq=(10,20,56,78,89)
num=int(input("enter a number:"))
if num in seq:
    print("found")
else:
    print("not found")
```

# 5.Write a program to find left shift and right shift values of a given number.

```python
num=int(input("enter a number:"))
shift=int(input("enter a shift value:"))
right=num>>shift
left=num<<shift
print("right shift:",right) 
print("left shift:",left)   
```
