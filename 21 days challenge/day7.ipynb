# Print a multiplication table

```python
num=int(input("enter the number:"))
for i in range(1,11):
    print(num, "x", i, "=", num*i)
```

# Check if its a Palindrome number

```python
word = input("Enter a word: ")

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
