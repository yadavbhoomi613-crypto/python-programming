# INDEXING (0-based)

```python
s = "Python"
print(s[0])   # 'P'
print(s[-1])  # 'n'

print(s[len(s)-1])
print(s[len(s)])
```

```python
s = "Python"
print(s[0])   # 'P'
print(s[-1])  # 'n'

print(s[len(s)-1])
```

# Common methods
append(x), extend(iterable), insert(i,x)  
remove(x), pop(i), clear()    
index(x), count(x)  
sort(), reverse()  

```python
L = [3, 1, 2]
L.append(5)         
print(L)
L.sort(reverse=True)            
print(L)
L.pop(1)
print(L)
```

# OPERATIONS

```python
squares = [x*x for x in range(1,6)]
evens = [x for x in range(10) if x%2==0]
pairs = [(x,y) for x in [1,2] for y in [3,4]]
print(squares)
print(evens)
print(pairs)
```
