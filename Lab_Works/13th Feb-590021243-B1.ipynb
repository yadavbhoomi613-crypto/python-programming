```python
class Toyota:
    def dummy_mileage_kmpl(self, engine_l: float, weight_kg: float, tyre_in: float) -> float:
        """
        Dummy mileage estimator in km/L.
        """
        mileage = 28 - 4 * engine_l - 0.006 * (weight_kg - 1000) - 0.7 * (tyre_in - 15)
        return max(5.0, mileage)


def city_highway_kmpl(engine_l: float, weight_kg: float, tyre_in: float):
    car = Toyota()
    base = car.dummy_mileage_kmpl(engine_l, weight_kg, tyre_in)
    city = 0.85 * base
    highway = 1.10 * base
    return base, city, highway


# Example usage:
if __name__ == "__main__":
    E, W, T = 1.5, 1200, 16
    base, city, highway = city_highway_kmpl(E, W, T)

    print(f"Inputs: engine={E}L, weight={W}kg, tyre={T}in")
    print(f"Base mileage   : {base:.2f} km/L")
    print(f"City mileage   : {city:.2f} km/L")
    print(f"Highway mileage: {highway:.2f} km/L")
```
