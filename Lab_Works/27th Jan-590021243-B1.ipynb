# PALINDROME NUMBERS IN PYTHON

```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

if num == rev:
    print("Palindrome number"
else:
    print("Not a palindrome number")
```

# ARMSTRONG NUMBER IN PYTHON

```python
num = int(input("Enter a number: "))
temp = num
digits = 0
sum = 0

t = num
while t > 0:
    digits += 1
    t //= 10
    
while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp //= 10

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
