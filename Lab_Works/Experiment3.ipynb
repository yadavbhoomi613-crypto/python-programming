### 1. Check whether a given number is divisible by 3 and 5 both.

```python
n = int(input("Enter number: "))

if n % 3 == 0 and n % 5 == 0:
    print("Divisible by 3 and 5")
else:
    print("Not divisible by both")
```

### 2. Check whether a given number is multiple of five or not.

```python
n = int(input("Enter number: "))

if n % 5 == 0:
    print("Multiple of 5")
else:
    print("Not multiple of 5")
```

### 3. Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.

```python
a = int(input("Enter first: "))
b = int(input("Enter second: "))

if a > b:
    print("Greatest:", a)
elif b > a:
    print("Greatest:", b)
else:
    print("Numbers are equal")
```

### 4. Find the greatest among three numbers assuming no two values are same. 

```python
a = int(input("Enter first number:"))
b = int(input("Enter second number:"))
c = int(input("Enter third number:"))
print("First number:",a)
print("Second number:",b)
print("Third number:",c)
if a > b and a > c:
    print("Greatest:", a)
elif b > a and b > c:
    print("Greatest:", b)
else:
    print("Greatest:", c)
```

### 5. Check whether the quadratic equation has real roots or imaginary roots.Display the roots.


```python
import math

a = float(input("a: "))
b = float(input("b: "))
c = float(input("c: "))
print(a,"x2",b,"x",c)
d = b*b - 4*a*c

if d > 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Real roots:", r1, r2)

elif d == 0:
    r = -b / (2*a)
    print("Equal roots:", r)

else:
    print("Imaginary roots")
```

### 6. Find whether a given year is a leap year or not.

```python
y = int(input("Enter year: "))
print(y)
if (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

### 7. Write a program which takes any date as input and display next date of the calendar e.g.
### I/P: day=20 month=9 year=2005
### O/P: day=21 month=9 year 2005

```python
day = int(input("Day: "))
month = int(input("Month: "))
year = int(input("Year: "))
print("Entered Date:",day,month,year)
day += 1

if day > 30:  
    day = 1
    month += 1

if month > 12:
    month = 1
    year += 1

print("Next date:", day, month, year)
```

### 8. Print the grade sheet of a student for the given range of cgpa. Scan marks of
### five subjects and calculate the percentage.
### CGPA=percentage/10
### CGPA range:
### 0 to 3.4 -> F
### 3.5 to 5.0->C+
### 5.1 to 6->B
### 6.1 to 7-> B+
### 7.1 to 8-> A
### 8.1 to 9->A+
### 9.1 to 10-> O (Outstanding)
### Sample Gradesheet
### Name: Rohit Sharma
### Roll Number: R17234512 SAPID: 50005673
### Sem: 1 Course: B.Tech. CSE AI&ML
### Subject name: Marks
### PDS: 70
### Python: 80
### Chemistry: 90
### English: 60
### Physics: 50
### Percentage: 70%
### CGPA:7.0
### Grade: A

```python
name = input("Name: ")

m1 = int(input("PDS: "))
m2 = int(input("Python: "))
m3 = int(input("Chemistry: "))
m4 = int(input("English: "))
m5 = int(input("Physics: "))

total = m1 + m2 + m3 + m4 + m5
percentage = total / 5
cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6:
    grade = "B"
elif cgpa <= 7:
    grade = "B+"
elif cgpa <= 8:
    grade = "A"
elif cgpa <= 9:
    grade = "A+"
else:
    grade = "O"

print("\n--- Grade Sheet ---")
print("Name:", name)
print("Percentage:", percentage)
print("CGPA:", round(cgpa,1))
print("Grade:", grade)
```
