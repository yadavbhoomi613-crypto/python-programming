### Q1) Find a factorial of given number.

```python
a=int(input("Enter a number"))
print("Entered Number:",a)
fact = 1

for i in range(1, a+1):
    fact *= i

print("Factorial =", fact)
```

### Q2) Find whether the given number is Armstrong number.

```python
num=int(input("Enter a number:"))
print("Entered number:",num)
temp = num
power = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** power
    temp //= 10

print("Armstrong" if sum == num else "Not Armstrong")
```

### Q3) Print Fibonacci series up to given term


```python
num=int(input("Enter a number:"))
print("Enter number:",num)
a, b = 0, 1

for _ in range(num):
    print(a, end=" ")
    a, b = b, a + b
```

### Q4) Write a program to find if given number is prime number or not.


```python
num=int(input("Enter a number"))
print("Entered number:",num) 
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

### Q5) Check whether given number is palindrome or not.

```python
num=int(input("Enter the number:"))
print("Entered number:",num)
temp = num
rev = 0

while temp > 0:
    rev = rev*10 + temp%10
    temp //= 10

print("Palindrome" if rev == num else "Not Palindrome")
```

### Q6) Write a program to print sum of digits


```python
num=int(input("Enter a number :"))
print("Entereed number:",num)
s = 0

while num > 0:
    s += num % 10
    num //= 10

print("Sum =", s)
```

### Q7) Count and print all numbers divisible by 5 or 7 between 1 to 100.

```python
count = 0
for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nTotal =", count)
```

### Q9) Print all prime numbers between 1 and 100.

```python

for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```
