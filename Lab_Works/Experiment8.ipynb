### 1. Add few names, one name in each row, in “name.txt file”.
### a. Count no of names
### b. Count all names starting with vowel
### c. Find longest name

```python
with open("name.txt", "w") as f:
    f.write("Aman\n")
    f.write("Riya\n")
    f.write("Ishita\n")
    f.write("Om\n")
    f.write("Uday\n")
```

a

```python
with open("name.txt", "r") as f:
    names = f.readlines()

print("Total names:", len(names))
```

b

```python
vowels = "AEIOUaeiou"

count = 0
with open("name.txt", "r") as f:
    for name in f:
        if name.strip()[0] in vowels:
            count += 1

print("Names starting with vowel:", count)
```

c

```python
with open("name.txt", "r") as f:
    names = [name.strip() for name in f]

longest = names[0]
for name in names:
    if len(name) > len(longest):
        longest = name

print("Longest name:", longest)
```

### 2. Store integers in a file.
### a. Find the max number
### b. Find average of all numbers
### c. Count number of numbers greater than 100


```python
with open("numbers.txt", "w") as f:
    f.write("50\n120\n300\n45\n150\n")
```

a

```python
with open("numbers.txt", "r") as f:
    nums = [int(x.strip()) for x in f]

print("Maximum number:", max(nums))
```

b

```python
print("Average:", sum(nums)/len(nums))
```

c

```python
count = 0
for n in nums:
    if n > 100:
        count += 1

print("Numbers greater than 100:", count)
```

### a. Display details of all cities

### c. Display sum of areas of all cities


a

```python
with open("city.txt", "r") as f:
    for line in f:
        print(line.strip())
```

b

```python
with open("city.txt", "r") as f:
    for line in f:
        city, pop, area = line.split()
        if float(pop) > 10:
            print(city)
```

c

```python
total_area = 0

with open("city.txt", "r") as f:
    for line in f:
        city, pop, area = line.split()
        total_area += float(area)

print("Total area:", total_area)
```

### 4. Input two values from user where the first line contains N, the number of test
### cases. The next N lines contain the space separated values of a and b. Perform
### integer division and print a/b. Handle exception in case of ZeroDivisionError or
### ValueError.
### Sample input
### 1 0
### 2 $
### 3 1
### Sample Output:
### Error Code: integer division or modulo by zero
### Error Code: invalid literal for int() with base 10: '$' 3

```python
n = int(input("Enter number of test cases: "))

for i in range(n):
    try:
        a, b = input().split()
        print(int(a) // int(b))
    except ZeroDivisionError as e:
        print("Error Code:", e)
    except ValueError as e:
        print("Error Code:", e)
```

### 5. Create multiple suitable exceptions for a file handling program.


```python
class FileEmptyError(Exception):
    pass

class InvalidDataError(Exception):
    pass

try:
    with open("data.txt", "r") as f:
        content = f.read()
        
        if not content:
            raise FileEmptyError("File is empty")
        
        if not content.isdigit():
            raise InvalidDataError("File contains invalid data")
        
        print("File content is valid")

except FileNotFoundError:
    print("File not found")
except FileEmptyError as e:
    print("Error:", e)
except InvalidDataError as e:
    print("Error:", e)
```
