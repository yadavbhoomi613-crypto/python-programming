# Q3

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) / 2
    return int(total - sum(nums))
```

```python
nums = [3, 0, 1]
print(missing_number(nums))
```

# Q2

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d
print(count_by_category([("sports","win"), ("sports","win"), ("tech","ai")]))
```

# Q4

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
    return list(seen)
print(add_unique(1))
print(add_unique(2))
```
